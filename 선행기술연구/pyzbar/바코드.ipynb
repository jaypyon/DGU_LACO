```python
import pyzbar.pyzbar as pyzbar
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("불티나.png")
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
decoded = pyzbar.decode(gray)
```

```python
decoded
```

```python
for d in decoded:
    print(d.data.decode('utf-8'))
```
